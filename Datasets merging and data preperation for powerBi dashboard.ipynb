## Datasets Merging
 Here all the Datasets (HCHO,Weather,Covid and demographics,anthropogenic activities) will be merged to get the final dataset

```python
weather=pd.read_csv(r"processed data\weather.csv")
cov_demo=pd.read_csv(r"processed data\covid_and_demographics.csv")
anthro=pd.read_csv(r"processed data\anthropogenic_effects.csv")
hcho=pd.read_csv(r"processed data\temp\all_fomaldehyde.csv")
```

```python
print(weather.shape)
weather.columns
```

```python
print(cov_demo.shape)
cov_demo.columns
```

```python
print(anthro.shape)
anthro.columns
```

```python
print(hcho.shape)
hcho.columns
```

```python
merged_df = pd.merge(hcho, weather, left_on=['Current_Date', 'Location'], right_on=['DATE', 'NAME'], how='inner')
merged_df = merged_df.drop(['DATE', 'NAME'], axis=1)
merged_df = pd.merge(merged_df, cov_demo, left_on=['Current_Date', 'Location'], right_on=['date', 'District'], how='inner')
merged_df = merged_df.drop(['date', 'District'], axis=1)
merged_df = pd.merge(merged_df, anthro, left_on=['Current_Date', 'Location'], right_on=['DATE', 'region'], how='inner')
merged_df = merged_df.drop(['DATE', 'region'], axis=1)
merged_df=merged_df.sort_values(by=['Location', 'Current_Date'])
```

```python
merged_df.columns
```

```python
try:
    # Convert 'Current_Date' to DateTime format
    merged_df['Next_Date'] = pd.to_datetime(merged_df['Next_Date'])
    print("Conversion to DateTime format successful.")
except ValueError as e:
    print(f"Error converting to DateTime format: {e}")
```

```python
try:
    # Convert 'Current_Date' to DateTime format
    merged_df['Current_Date'] = pd.to_datetime(merged_df['Current_Date'])
    print("Conversion to DateTime format successful.")
except ValueError as e:
    print(f"Error converting to DateTime format: {e}")
```

```python
merged_df=merged_df[['Current_Date', 'Location','LATITUDE', 'LONGITUDE', 'ELEVATION', 'Sea Proximity (approximate-km)', 
       'Sea Proximity Categorised', 'Climate Zone','Area(km^2)', 'Population', 'Population_density','new','total',
       'Lockdown', 'carbonmonoxide_average', 'nitrogendioxide_average','ozone_average','PRCP', 'TAVG', 'TMAX', 'TMIN',
       'HCHO_reading']]
merged_df.columns
```

## Dataset Preperation for Power BI dashboard

Here the initial transformations need to be done in the dataset for powerBI dashboard will be done.

```python
df=merged_df.copy()
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['Lockdown','TMAX','TMIN'], inplace=True)
df.columns
```

```python
df = df.rename(columns={'PRCP':'Precipitation','TAVG': 'Average Temperature', 
                        'carbonmonoxide_average': 'Average Carbonmonoxide', 
                        'ozone_average':'Average Ozone','nitrogendioxide_average':'Average Nitrogendioxide',
                       'total':'Total Cases','new':'Daily Cases'})
df['HCHO Reading']=df['HCHO_reading']
```

```python
df.columns
```

```python
var = ['HCHO Reading','Precipitation', 'Average Temperature', 'Average Carbonmonoxide', 'Average Nitrogendioxide'
       , 'Average Ozone']
id_cols = df.columns.difference(var).tolist()
melted_df = pd.melt(df, id_vars=id_cols, value_vars=var, var_name='Weather_and_Anthro', value_name='Reading')

var2=['Total Cases', 'Daily Cases']
id_cols2=melted_df.columns.difference(var2).tolist()
melted_df = pd.melt(melted_df, id_vars=id_cols2, value_vars=var2, var_name='Covid Data', value_name='Reading2')

melted_df.sort_values(by=['Current_Date','Location' , 'Weather_and_Anthro'],inplace=True)
melted_df=melted_df[['Current_Date', 'Location', 'LATITUDE',
       'LONGITUDE', 'ELEVATION', 'Sea Proximity (approximate-km)','Sea Proximity Categorised','Area(km^2)', 'Population', 'Population_density',
        'Climate Zone','Covid Data', 'Reading2','Weather_and_Anthro','Reading','HCHO_reading']]
print(melted_df)
```
