```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
import numpy as np
# Set the style to 'whitegrid'
plt.style.use('seaborn-whitegrid')
```

```python
df=pd.read_csv('raw_data/weather-data.csv',header=None)
print(df.shape)
df.head(10)
```

```python
# Extract the first row
header_row = df.iloc[0]
# Assign the first row as the header
df.columns = header_row
# Remove the first row from the DataFrame
df = df[1:]

df.head(10)
```

```python
df.drop(columns=['STATION', 'PRCP_ATTRIBUTES', 'TAVG_ATTRIBUTES', 'TMAX_ATTRIBUTES', 'TMIN_ATTRIBUTES'], inplace=True)
df.head(10)
```

```python
df['NAME'].value_counts()
```

```python
try:
    # Convert 'Current_Date' to DateTime format
    df['DATE'] = pd.to_datetime(df['DATE'])
    print("Conversion to DateTime format successful.")
except ValueError as e:
    print(f"Error converting to DateTime format: {e}")
```

```python
cmb= df[df['NAME']=='COLOMBO, CE']
print(cmb.shape)
cmb.head(10)
```

```python
# Step 1: Check Time Range
min_date = cmb['DATE'].min()
max_date = cmb['DATE'].max()
print("Time Range: From", min_date, "to", max_date)

# Step 2: Generate Time Range
complete_time_range = pd.date_range(start=min_date, end=max_date, freq='D')  
# Step 3: Identify Missing Values
missing_values = complete_time_range[~complete_time_range.isin(cmb['DATE'])]

if len(missing_values) == 0:
    print("No missing dates found in the time range.")
else:
    print("Missing values found:")
    print('Count: ',len(missing_values))
    print(missing_values)
```

```python
# Step 4: Create DataFrame for missing dates
missing_df = pd.DataFrame({'DATE': missing_values,
                           'NAME': ['KURUNEGALA, CE']*len(missing_values),
                           'LATITUDE':['7.467']*len(missing_values), 
                           'LONGITUDE':['80.367']*len(missing_values),
                           'ELEVATION':['116.0']*len(missing_values)})

# Step 5: Concatenate missing dates DataFrame with existing DataFrame
cmb_with_missing_dates = pd.concat([cmb, missing_df], ignore_index=True)

# Sort the DataFrame by 'DATE' column
cmb_with_missing_dates = cmb_with_missing_dates.sort_values(by='DATE')

print("DataFrame with missing dates added:")
cmb_with_missing_dates[cmb_with_missing_dates['DATE'].isin(missing_values)]
```

```python
cmb_with_missing_dates['PRCP'] = pd.to_numeric(cmb_with_missing_dates['PRCP'], errors='coerce')
cmb_with_missing_dates['TAVG'] = pd.to_numeric(cmb_with_missing_dates['TAVG'], errors='coerce')
cmb_with_missing_dates['TMIN'] = pd.to_numeric(cmb_with_missing_dates['TMIN'], errors='coerce')
cmb_with_missing_dates['TMAX'] = pd.to_numeric(cmb_with_missing_dates['TMAX'], errors='coerce')
```

```python
# Visualize missing values using a heatmap
plt.figure(figsize=(6, 20))
sns.heatmap(cmb.isnull(), cbar=False, cmap='inferno')
plt.title('Disribution of Missing Values in Weather Data')
plt.xlabel('Columns')
plt.ylabel('Date')
plt.show()
```
